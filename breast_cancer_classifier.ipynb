```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
```

## Loading dataset

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
dataset
```

# Creating Dataframe

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df
```

# Adding target column to dataframe

```python
df['label'] = dataset.target
```

```python
df.shape
```

```python
df.info()
```

# Statistical measure of the data 

```python
df.describe()
```

## Splitting df into input and output

```python
X = df.iloc[:,:-1]
```

```python
y = df['label']
```

```python
X
```

```python
y
```

## Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=14)
```

```python
# Checking the shape
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

## Standardizing the X_train and X_test

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_trans = scaler.fit_transform(X_train)
```

```python
X_test_trans = scaler.transform(X_test)
```

```python
X_train_trans
```

```python
X_test_trans
```

## Creating Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train_trans,y_train)
```

```python
# Checking the coefficients

log_reg_model.coef_
```

```python
# Checking the intercept

log_reg_model.intercept_
```

## Evaluating the model accuracy on train data

```python
y_train_pred = log_reg_model.predict(X_train_trans)

temp_df = pd.DataFrame({'Actual': y_train, 'Predicted': y_train_pred})

#temp_df.head()

sns.histplot(y_train, color='blue', alpha=0.5)
sns.histplot(y_train_pred, color='red', alpha=0.5)

from sklearn.metrics import accuracy_score
accu_logr = accuracy_score(y_train, y_train_pred)*100
print('accuracy: ', accu_logr)
```

## Evaluating the model accuracy on test data

```python
y_test_pred = log_reg_model.predict(X_test_trans)

temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

#temp_df.head()

sns.histplot(y_test, color='blue', alpha=0.5)
sns.histplot(y_test_pred, color='red', alpha=0.5)

from sklearn.metrics import accuracy_score
accu_logr=accuracy_score(y_test, y_test_pred)*100
print('accuracy: ', accu_logr)
```

```python
#!pip install tensorflow
```

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
    
])
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
history = model.fit(X_train_trans,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel("Epoch")

plt.legend(['training data', 'validation data'], loc = 'lower right' )
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel("Epoch")

plt.legend(['training data', 'validation data'], loc = 'upper right' )
```

## Checking the accuracy on test data

```python
loss, accuracy = model.evaluate(X_test_trans,y_test)
print(accuracy)
```

## Building the predictive system

```python
input_data = (17.99, 10.38, 122.8, 1001.0, 0.1184, 0.2776, 0.3001, 0.1471, 0.2419, 0.07871, 1.095, 0.9053, 8.589, 153.4, 
              0.006399, 0.04904, 0.05373, 0.01587, 0.03003, 0.006193, 25.38, 17.33, 184.6, 2019.0, 0.1622, 0.6656, 0.7119,
              0.2654, 0.4601, 0.1189)

# changing the input data as numpy array

input_data_as_array = np.asarray(input_data)

# reshaping the array

input_data_reshaped = input_data_as_array.reshape(1,-1)

# Scaling the reshaped input data

input_data_std = scaler.transform(input_data_reshaped)

# Predicting

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if prediction_label[0] == 0:
    print("The tumor is MALIGNANT")
else:
    print("The tumor is BENIGN")
```
